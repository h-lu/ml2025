# 机器学习(Python) 
## 2025年春季学期
## 课程简介

本课程面向经管学生，通过项目驱动的方式，帮助学生掌握机器学习和生成式AI的基本原理与商业应用能力。课程分为传统机器学习和生成式AI两大模块，每个模块都包含理论讲解和实战项目。

## 课程目标

- 掌握机器学习基础知识和主要算法
- 熟练使用AI辅助工具进行开发
- 培养解决实际商业问题的能力
- 了解生成式AI的最新发展与应用

## 课程安排

### 第一部分：机器学习基础 (1-9周)

#### 第1-3周：分类问题
- 机器学习基础概念
- 数据预处理与特征工程
- 常用分类算法
- 模型评估方法
- 项目1：分类问题实战

#### 第4-6周：回归分析
- 回归模型原理
- 模型优化与调参
- 集成学习方法
- 项目2：回归问题实战

#### 第7-9周：无监督学习
- 聚类算法
- 降维技术
- 异常检测
- 项目3：无监督学习实战

### 第二部分：生成式AI应用 (10-16周)

#### 第10-11周：大语言模型应用
- Prompt工程基础
- LLM商业应用实践
- Agent开发入门

#### 第12-13周：多模态AI
- 文生图技术应用
- 视频生成基础
- 项目4：Agent开发实战

#### 第14-16周：综合项目
- 商业创新方案设计
- 原型开发
- 项目答辩与展示

## 考核方式

### 平时表现 (10%)
- 课堂参与
- 小组讨论
### 项目考核 (60%)
1. 分类项目 (10%)
2. 回归项目 (10%)
3. 无监督学习项目 (10%)
4. Agent开发项目 (10%)
5. 综合项目 (20%)
### 期末考试 (30%)
- 闭卷考试
- 基本概念与原理
- 应用场景与案例分析
